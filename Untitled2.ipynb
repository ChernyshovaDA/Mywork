```python
import numpy as np
import matplotlib.pyplot as plt
s = np.loadtxt("http://scipy-lectures.org/_downloads/populations.txt")
```

Exercise on Numpy and Matplotlib

The data in populations.txt describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Get the data with np.loadtxt and do the following exercises without for-loops.

Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
plt.plot(s[:,0], s[:,1], label='hare', c='red')
plt.plot(s[:,0], s[:,2], label='lynx', c ='yellow')
plt.plot(s[:,0], s[:,3], label='carrot', c ='green')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
plt.bar(s[:,0], s[:,1]/(s[:,1]+s[:,2]), label='hares')
plt.bar(s[:,0], s[:,2]/(s[:,1]+s[:,2]), bottom =  s[:,1]/(s[:,1]+s[:,2]), label='lynxes')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
fig, ax=plt.subplots(1,3)
ax[0].hist(s[:,1], label='hares', bins=len(s[:,0]))
ax[0].axvline(np.mean(s[:,1]), c = 'red')
ax[1].hist(s[:,2], label='lynxes', bins=len(s[:,0]))
ax[1].axvline(np.mean(s[:,2]), c = 'red')
ax[2].hist(s[:,3], label='carrots', bins=len(s[:,0]))
ax[2].axvline(np.mean(s[:,3]), c ='red')
plt.legend()
print("mean and std of hares ",np.mean(s[:,1]),",",np.std(s[:,1]))
print("mean and std of lynxes ",np.mean(s[:,2]),",",np.std(s[:,2]))
print("mean and std of carrots",np.mean(s[:,3]),",",np.std(s[:,3]))
plt.show()
```

Find which year each species had the largest population. Mark these years in the plot of populations.

```python
plt.plot(s[:,0], s[:,1], label='hare', c='red')
plt.vlines(s[np.argmax(s[:,1]),0], 0, np.max(s[:,1]))
plt.plot(s[:,0], s[:,2], label='lynx', c ='yellow')
plt.vlines(s[np.argmax(s[:,2]),0],0, np.max(s[:,2]))
plt.plot(s[:,0], s[:,3], label='carrot', c ='green')
plt.vlines(s[np.argmax(s[:,3]),0],0, np.max(s[:,3]))
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing ratio of "H" and "L" values obtained.

```python
print (np.where(s[:,1]<s[:,2], 'L','H'))
plt.pie([np.sum(s[:,1]>s[:,2]), np.sum(s[:,1]<s[:,2])], labels = ['H','L'], autopct='%1.1f%%')
plt.show()
```

Find which years any of the populations is above 50000. Show time periods where populations are above 50000 in the plot of populations.

```python
plt.plot(s[:,0], s[:,1], label='hare', c='red')
plt.vlines(s[np.where(s[:,1]>50000),0], 0, s[np.where(s[:,1]>50000),1] )
plt.plot(s[:,0], s[:,2], label='lynx', c ='yellow')
plt.vlines(s[np.where(s[:,2]>50000),0],0, s[np.where(s[:,2]>50000),2])
plt.plot(s[:,0], s[:,3], label='carrot', c ='green')
plt.vlines(s[np.where(s[:,3]>50000),0],0, s[np.where(s[:,3]>50000),3])
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

Find the top 2 years for each species when they had the lowest populations.

```python
print('Hares:', s[np.argsort(s[:,1])[0], 0], s[np.argsort(s[:,1])[1], 0] )
print('Lynxes:', s[np.argsort(s[:,2])[0], 0], s[np.argsort(s[:,2])[1], 0] )
print('Carrots:', s[np.argsort(s[:,3])[0], 0], s[np.argsort(s[:,3])[1], 0] )
```

Plot the change in hare population and the number of lynxes. Find the correlation coefficient.

```python
print('Correlation coefficient:', np.corrcoef(np.diff(s[:,1]),np.diff(s[:,2]))[0,1])
plt.plot(s[:,0], np.insert(np.diff(s[:,1]),0, 0), label='hare', c='red')
plt.plot(s[:,0], np.insert(np.diff(s[:,2]),0,0), label='lynx', c ='yellow')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

Show population of hares vs carrots and hares vs lynxes.

```python
fig, ax=plt.subplots(1,2)
fig.suptitle('hares vs carrots')
ax[0].plot(s[:,1], label='hares')
ax[1].plot(s[:,3], label='carrots')
fig, ax=plt.subplots(1,2)
fig.suptitle('hares vs lynxes')
ax[0].plot(s[:,1], label='hares')
ax[1].plot(s[:,2], label='lynxes')
plt.legend()
plt.show()
```
